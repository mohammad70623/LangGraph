```python
from typing_extensions import TypedDict
from langgraph.graph import StateGraph, START, END
from langchain_groq import ChatGroq
from dotenv import load_dotenv
```

```python
load_dotenv() 
subgraph_model = ChatGroq(model="llama-3.3-70b-versatile")
parent_model = ChatGroq(model="openai/gpt-oss-safeguard-20b")
```

```python
class ParentState(TypedDict):

    question: str
    answer_eng: str
    answer_ban: str
```

```python
def translate_text(state: ParentState):

    prompt = f"""
Translate the following text into Bangla. 
Ensure the translation is natural and clear without adding extra information.

Text:
{state["input_text"]}
""".strip()
    
    translated_text = subgraph_model.invoke(prompt).content

    return {'translated_text': translated_text}
```
